```python
def hill_huntington_apportionment(df, total_representatives=435):
    """
    Calculates apportionment using the Hill-Huntington method
    """
    # 1 representative per state
    df['Apportionment'] = 1
    remaining_representatives = total_representatives - len(df)

    # Calculate initial priority for each state
    df['Priority'] = df['Population'] / ((df['Apportionment'] * (df['Apportionment'] + 1))**0.5)

    # assign remaining representatives
    for _ in range(remaining_representatives):
        # state with the highest priority
        max_priority_state_index = df['Priority'].idxmax()

        # additional representative
        df.loc[max_priority_state_index, 'Apportionment'] += 1

        # Recalculate the priority
        state_pop = df.loc[max_priority_state_index, 'Population']
        state_apportionment = df.loc[max_priority_state_index, 'Apportionment']
        df.loc[max_priority_state_index, 'Priority'] = state_pop / ((state_apportionment * (state_apportionment + 1))**0.5)

    return df[['State', 'Population', 'Apportionment']]

# Ensure the 'Population' column
df_states['Population'] = df_states['Population'].str.replace(',', '').astype(int)

# Calculate the apportionment
df_apportioned = hill_huntington_apportionment(df_states.copy())

# Display the calculated apportionment
display(df_apportioned.head())
```
