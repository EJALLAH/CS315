```
import time
import random
```

```
def time_sorting_algorithm(sort_func, n, data_type='random'):
    if data_type == 'random':
        data = [random.randint(0, n) for _ in range(n)]
    elif data_type == 'sorted':
        data = list(range(n))
    else:
        raise ValueError("data_type has to be 'random' or 'sorted'")

    start_time = time.time()
    # Creating a copy
    sorted_data = sort_func(data.copy())
    end_time = time.time()

    return end_time - start_time
```

```
import pandas as pd
#inputing different size numbers
input_sizes = [1000, 2000, 4000, 8000, 16000, 32000]
sorting_algorithms = {
    "Selection Sort": selection_sort,
    "Insertion Sort": insertion_sort,
    "Merge Sort": merge_sort,
    "Quicksort": quicksort,
}
#Creating the time result for each sorts
timing_results_random = []

for name, sort_func in sorting_algorithms.items():
    print(f"Timing {name} on random data...")
    for n in input_sizes:
        try:
            time_taken = time_sorting_algorithm(sort_func, n, data_type='random')
            timing_results_random.append({"Algorithm": name, "Input Size": n, "Time (s)": time_taken})
            print(f"  N={n}: {time_taken:.6f} seconds")
        except RecursionError:
            print(f"  N={n}: Recursion depth exceeded for {name}")
            timing_results_random.append({"Algorithm": name, "Input Size": n, "Time (s)": float('inf')}) # Indicate failure

timing_df_random = pd.DataFrame(timing_results_random)
display(timing_df_random)
```
